```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x_df = pd.read_csv('x_train.csv')
y_df = pd.read_csv('y_train.csv')
```

```python
plot_data = x_df.groupby("Sample_Number").first().reset_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(plot_data["DAT"], plot_data["fruit_count"], 'o', label="fruit_count", color='red', alpha=0.6)
plt.plot(plot_data["DAT"], plot_data["flower_count"], 'o', label="flower_count", color='blue', alpha=0.6)

plt.xlabel("DAT")
plt.ylabel("Count")
plt.title("Fruit and Flower Count per DAT")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sample_Number   (10     )
plot_data = x_df.groupby("Sample_Number").first().reset_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(plot_data["DAT"], plot_data["plant_height"], 'o', color='green', label="plant_height")

plt.xlabel("DAT")
plt.ylabel("Plant Height (cm)")
plt.title("Plant Height per DAT")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
